```python
import networkx as nx
import numpy as np
from matplotlib import pyplot as plt
```

```python
edges = np.array([[1, 2], [2, 3], [3, 4], [1, 3], [5, 3]])
```

```python
edges
```

```python
gr = nx.Graph()
gr.add_edges_from(edges)
```

```python
plt.figure(figsize=(10, 10))
pos=nx.spring_layout(gr) # positions for all nodes

nx.draw_networkx_nodes(gr,pos,
                       node_color='b',
                       node_size=400,
                   alpha=0.8)

nx.draw_networkx_edges(gr,pos,width=1.0,alpha=0.5)
plt.show()
```

Let construct probability matrix of that graph, assuming that probabilities of transition are equal.

```python
adj = np.array(nx.adjacency_matrix(gr).todense())
```

```python
probability_matrix = adj / np.sum(adj, axis=1)
```

```python
starting_probability = np.ones(adj.shape[0]) / adj.shape[0]
```

```python
distribution = probability_matrix @ starting_probability
```

```python
plt.figure(figsize=(10, 10))
pos=nx.spring_layout(gr) # positions for all nodes

nx.draw_networkx_nodes(gr,pos,
                       node_color='b',
                       node_size=distribution * 7000,
                   alpha=0.8)

nx.draw_networkx_edges(gr,pos,width=1.0,alpha=0.5)
plt.show()
```
